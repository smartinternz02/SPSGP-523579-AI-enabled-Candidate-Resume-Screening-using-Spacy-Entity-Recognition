```python
#Task - 1: Creating a pandas DataFrame with random values using NumPy:

import pandas as pd
import numpy as np

# Create a numpy array with random values
np.random.seed(42)
data = np.random.randn(4, 4)

# Create a DataFrame using the numpy array
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
#Task - 2: Renaming the column names of the 'df' DataFrame:

df.columns = ['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4']

# Print the DataFrame with renamed column names
print(df)
```

```python
#Task - 3: Finding the descriptive statistics of the 'df' DataFrame:

statistics = df.describe()

# Print the descriptive statistics
print(statistics)
```

```python
#Task - 4: Checking for null values and data types of columns in the 'df' DataFrame:

null_values = df.isnull().sum()
data_types = df.dtypes

# Print the null values and data types
print("Null Values:")
print(null_values)
print("\nData Types:")
print(data_types)
```

```python
#Task - 5: Displaying 'Random value 2' & 'Random value 3' columns using location and index location methods:

# Using location method (using column names)
random_values_loc = df.loc[:, ['Random value 2', 'Random value 3']]

# Using index location method (using column indices)
random_values_iloc = df.iloc[:, [1, 2]]

# Print the selected columns
print("Random value 2 & Random value 3 (using location method):\n", random_values_loc)
print("\nRandom value 2 & Random value 3 (using index location method):\n", random_values_iloc)
```
